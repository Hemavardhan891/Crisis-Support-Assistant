```python
%%writefile app.py

# ============================================================================
# CRISIS SUPPORT ASSISTANT - GOOGLE ADK (GEMINI + STREAMLIT VERSION)
# ============================================================================

# IMPORTS
import os
import json
from typing import Dict, Any, List, Callable
from dataclasses import dataclass
import streamlit as st
from google import genai # The Google Gen AI SDK
from google.genai import types # Use types for function definitions and config

# --- Tool Functions (UNCHANGED) ---

@dataclass
class Tool:
    """Tool definition following ADK pattern"""
    name: str
    description: str
    function: Callable
    parameters: Dict[str, Any]

def get_breathing_exercise(duration: str = "short") -> dict:
    # ... (function body is UNCHANGED) ...
    exercises = {
        "short": {
            "name": "4-7-8 Breathing Technique",
            "duration": "1-2 minutes",
            "steps": [
                "Sit comfortably with back straight",
                "Exhale completely through mouth",
                "Inhale through nose for 4 counts",
                "Hold breath for 7 counts",
                "Exhale through mouth for 8 counts",
                "Repeat 3-4 times"
            ],
            "benefits": "Reduces anxiety quickly, calms nervous system"
        },
        "medium": {
            "name": "Box Breathing (Navy SEAL technique)",
            "duration": "3-5 minutes",
            "steps": [
                "Sit upright, feet flat on floor",
                "Breathe in for 4 counts",
                "Hold for 4 counts",
                "Breathe out for 4 counts",
                "Hold empty for 4 counts",
                "Repeat 10-15 times"
            ],
            "benefits": "Excellent for stress management and focus"
        },
        "long": {
            "name": "Progressive Relaxation Breathing",
            "duration": "5-10 minutes",
            "steps": [
                "Lie down comfortably",
                "Take 5 deep breaths",
                "Tense and relax each muscle group",
                "Start from toes, move to head",
                "Release tension with each exhale",
                "End with 5 peaceful breaths"
            ],
            "benefits": "Deep relaxation for severe anxiety"
        }
    }
    return {"success": True, "exercise": exercises.get(duration, exercises["short"])}

def get_grounding_technique(situation: str = "general") -> dict:
    # ... (function body is UNCHANGED) ...
    techniques = {
        "panic_attack": {
            "name": "5-4-3-2-1 Sensory Grounding",
            "description": "Interrupt panic by engaging all senses",
            "steps": [
                "Name 5 things you can SEE",
                "Name 4 things you can TOUCH",
                "Name 3 things you can HEAR",
                "Name 2 things you can SMELL",
                "Name 1 thing you can TASTE"
            ],
            "why": "Redirects attention from panic to safe reality"
        },
        "anxiety": {
            "name": "Physical Grounding",
            "description": "Use physical sensations to interrupt anxiety",
            "steps": [
                "Plant feet firmly on ground",
                "Press hands together hard (10 sec)",
                "Hold cold object (ice, water)",
                "Stretch body slowly",
                "Splash cold water on face"
            ],
            "why": "Physical sensations interrupt anxious thoughts"
        },
        "dissociation": {
            "name": "Reality Orientation",
            "description": "Reconnect mind and body",
            "steps": [
                "Say your full name out loud",
                "State today's date and location",
                "Touch different textures",
                "Describe surroundings in detail",
                "Name objects by color"
            ],
            "why": "Verbal and tactile awareness reconnects fragmented self"
        },
        "general": {
            "name": "Quick 54321 Method",
            "description": "Fast grounding for any situation",
            "steps": [
                "5 things you see",
                "4 things you touch",
                "3 things you hear",
                "2 things you smell",
                "1 thing you taste"
            ],
            "why": "Immediate present moment awareness"
        }
    }
    return {"success": True, "technique": techniques.get(situation, techniques["general"]), "reminder": "You are safe. This feeling will pass."}

def get_emergency_resources(location: str = "india") -> dict:
    # ... (function body is UNCHANGED) ...
    resources = {
        "india": {
            "country": "India",
            "emergency": "112",
            "hotlines": [
                {"name": "AASRA Suicide Prevention", "number": "+91-9820466726", "hours": "24/7"},
                {"name": "Vandrevala Foundation", "number": "1860-2662-345", "hours": "24/7"},
                {"name": "iCall Psychosocial Helpline", "number": "+91-9152987821", "hours": "Mon-Sat 8AM-10PM"},
                {"name": "Snehi", "number": "+91-9311767771", "hours": "24/7"},
                {"name": "Mann Talks", "number": "+91-8686139139", "hours": "24/7"}
            ]
        },
        "usa": {
            "country": "United States",
            "emergency": "911",
            "hotlines": [
                {"name": "988 Suicide & Crisis Lifeline", "number": "988", "hours": "24/7"},
                {"name": "Crisis Text Line", "number": "Text HOME to 741741", "hours": "24/7"},
                {"name": "SAMHSA Helpline", "number": "1-800-662-4357", "hours": "24/7"}
            ]
        },
        "uk": {
            "country": "United Kingdom",
            "emergency": "999",
            "hotlines": [
                {"name": "Samaritans", "number": "116 123", "hours": "24/7"},
                {"name": "Shout Crisis Text", "number": "Text SHOUT to 85258", "hours": "24/7"}
            ]
        }
    }
    return {"success": True, "resources": resources.get(location, resources["india"]), "urgent_warning": "‚ö†Ô∏è IF IN IMMEDIATE DANGER, CALL EMERGENCY SERVICES NOW"}

def get_coping_strategy(issue: str) -> dict:
    # ... (function body is UNCHANGED) ...
    strategies = {
        "anxiety": {
            "immediate": [
                "Practice 4-7-8 breathing",
                "Use 5-4-3-2-1 grounding",
                "Take a short walk",
                "Listen to calming music"
            ],
            "short_term": [
                "Journal anxious thoughts",
                "Challenge catastrophic thinking",
                "Limit caffeine intake",
                "Connect with a friend"
            ],
            "long_term": [
                "Consider CBT therapy (70-80% success rate)",
                "Regular exercise (3-5x/week)",
                "Daily meditation practice",
                "Maintain sleep schedule"
            ]
        },
        "panic": {
            "during_attack": [
                "Remember: This will pass, you are safe",
                "Use 5-4-3-2-1 grounding NOW",
                "Slow, deep breathing",
                "Don't fight the feelings"
            ],
            "after_attack": [
                "Rest and be gentle with yourself",
                "Drink water, eat something light",
                "Talk to someone you trust",
                "Avoid caffeine today"
            ],
            "prevention": [
                "Learn your panic triggers",
                "Practice breathing exercises daily",
                "Consider professional help",
                "Healthy sleep and diet"
            ]
        },
        "depression": {
            "immediate": [
                "Reach out to one person (text/call)",
                "One small self-care task",
                "Step outside for 5 minutes",
                "Drink water and eat something"
            ],
            "building_momentum": [
                "Set one tiny goal per day",
                "Move your body gently",
                "Challenge negative self-talk",
                "Keep gratitude journal"
            ],
            "professional_help": [
                "Talk to your doctor",
                "Consider therapy (CBT/IPT effective)",
                "Join support group",
                "Be patient with recovery"
            ]
        },
        "stress": {
            "quick_relief": [
                "Take 10 deep breaths",
                "5-minute body scan meditation",
                "Brief walk or stretch",
                "Write down what's stressing you"
            ],
            "management": [
                "Prioritize tasks (urgent/important)",
                "Practice saying 'no'",
                "Schedule regular breaks",
                "Connect with supportive people"
            ]
        }
    }
    return {"success": True, "strategies": strategies.get(issue, strategies["anxiety"]), "note": "Professional help is recommended for persistent issues"}

# ============================================================================
# TOOL EXECUTION MAPPER (UNCHANGED)
# ============================================================================

TOOL_FUNCTIONS = {
    "get_breathing_exercise": get_breathing_exercise,
    "get_grounding_technique": get_grounding_technique,
    "get_emergency_resources": get_emergency_resources,
    "get_coping_strategy": get_coping_strategy
}

# ============================================================================
# CRISIS ASSISTANT AGENT CLASS (REFRACTORED FOR GEMINI API)
# ============================================================================

class CrisisAssistantAgent:
    """Crisis Support Agent using ADK architectural pattern with Gemini API"""

    def __init__(self, api_key: str):
        # 1. Initialize Gemini Client
        # The client automatically picks up the API key from the environment variable (GEMINI_API_KEY)
        self.client = genai.Client(api_key=api_key)
        self.model_name = "gemini-2.5-flash" # Use a modern Gemini model

        # 2. Get the Chat Session instance, which handles conversation history
        self.chat = self.client.chats.create(
            model=self.model_name,
            config=types.GenerateContentConfig(
                system_instruction=self._get_system_prompt(),
                # 3. Use the Python functions directly as tools (Gemini SDK handles schema generation)
                tools=list(TOOL_FUNCTIONS.values())
            )
        )
        self.tools_map = TOOL_FUNCTIONS
        self.history_display = [] # Separate history for Streamlit display only

    def _get_system_prompt(self):
        return """You are a warm, empathetic, professional crisis support assistant.

CORE PRINCIPLES:
- Prioritize safety and wellbeing above all
- Be compassionate, non-judgmental, and validating
- Use tools proactively to provide concrete help
- Keep responses conversational and meaningful (2-4 paragraphs)
- Always encourage professional help for serious issues

WHEN TO USE TOOLS:
1. get_breathing_exercise - User mentions anxiety, panic, or needs calming
2. get_grounding_technique - User is panicking, dissociating, or overwhelmed
3. get_emergency_resources - CRITICAL: mentions self-harm, suicide, or severe crisis
4. get_coping_strategy - User needs strategies for ongoing challenges

COMMUNICATION STYLE:
- Warm and supportive like a caring, trained friend
- Validate feelings: "It makes sense you feel this way"
- Normalize struggles: "Many people experience this"
- Empower: "You're taking a brave step by reaching out"
- Be realistic: "Healing takes time, and that's okay"

IMPORTANT: You're a support tool, NOT a replacement for professional care."""


    def process_message(self, user_message: str) -> tuple[str, List[str]]:
        """
        Process user message with agentic tool use using Gemini Chat.
        """
        tools_used = []

        try:
            # Send message to the chat session
            response = self.chat.send_message(user_message)

            # Agentic Loop - Handle tool calls until a text response is received
            while response.function_calls:

                function_calls = response.function_calls
                tool_responses = []

                for function_call in function_calls:
                    tool_name = function_call.name
                    tool_args = dict(function_call.args)

                    # Execute tool
                    tool_func = self.tools_map.get(tool_name)
                    if tool_func:
                        result = tool_func(**tool_args)
                        tools_used.append(tool_name)

                        # Create the FunctionResponse object to send back to the model
                        tool_responses.append(
                            types.Part.from_function_response(
                                name=tool_name,
                                response={"result": result}
                            )
                        )

                # Send tool results back to the model
                response = self.chat.send_message(tool_responses)

            # Final text response
            final_text = response.text

            return final_text, tools_used

        except Exception as e:
            error_msg = f"I apologize, I'm having trouble connecting to the Gemini API. Please try again or contact emergency services if urgent.\n\nError: {str(e)}"
            return error_msg, []

    def reset(self):
        """Reset conversation history by creating a new chat session."""
        self.chat = self.client.chats.create(
            model=self.model_name,
            config=types.GenerateContentConfig(
                system_instruction=self._get_system_prompt(),
                tools=list(TOOL_FUNCTIONS.values())
            )
        )


# ============================================================================
# INITIALIZE AGENT & STREAMLIT UI (MODIFIED TO USE GEMINI AGENT)
# ============================================================================

if 'agent' not in st.session_state:
    try:
        # Client will look for GEMINI_API_KEY or GOOGLE_API_KEY
        api_key_env = os.environ.get('GEMINI_API_KEY')
        if not api_key_env:
             st.error("GEMINI_API_KEY environment variable is not set. Please run Cell 2 first.")
             st.stop()

        st.session_state['agent'] = CrisisAssistantAgent(api_key=api_key_env)
        st.session_state['messages'] = []
    except Exception as e:
        st.error(f"Error initializing Gemini agent: {e}")
        st.stop()

# --- Streamlit UI Setup (Essentially UNCHANGED) ---

st.set_page_config(
    page_title="üíú Crisis Support Assistant (Gemini)",
    layout="wide",
    initial_sidebar_state="collapsed"
)

# --- Emergency Banner ---
emergency_html = """
<div style='background: #fee2e2; border: 3px solid #dc2626; border-radius: 12px; padding: 15px; margin-bottom: 20px;'>
    <h3 style='color: #991b1b; margin: 0 0 10px 0;'>üö® EMERGENCY RESOURCES - INDIA</h3>
    <p style='margin: 5px 0;'><strong>Emergency Services:</strong> <span style='font-size: 20px; font-weight: bold; color: #dc2626;'>112</span></p>
    <p style='margin: 5px 0;'><strong>AASRA Suicide Prevention:</strong> +91-9820466726 (24/7)</p>
    <p style='margin-top: 10px; text-align: center; font-style: italic; color: #7f1d1d;'><strong>‚ö†Ô∏è If you're in immediate danger, please call emergency services now.</strong></p>
</div>
"""
st.markdown(emergency_html, unsafe_allow_html=True)
# --- End Emergency Banner ---

st.title("Crisis Support Assistant ")
st.markdown("---")

# --- Quick Responses ---
st.subheader("üéØ Quick Responses")
col1, col2, col3, col4 = st.columns(4)

def handle_quick_response(prompt):
    st.session_state.prompt_input = prompt
    st.session_state.run_chat = True
    st.rerun()

with col1:
    if st.button("üò∞ I'm feeling anxious", use_container_width=True):
        handle_quick_response("I'm feeling anxious")
with col2:
    if st.button("ü´Å I need breathing exercises", use_container_width=True):
        handle_quick_response("I need breathing exercises")
with col3:
    if st.button("üò® I'm having a panic attack", use_container_width=True):
        handle_quick_response("I'm having a panic attack")
with col4:
    if st.button("üí≠ I need to talk", use_container_width=True):
        handle_quick_response("I need someone to talk to")

st.markdown("---")
# --- End Quick Responses ---


# --- Chat Interface Logic ---

for message in st.session_state.messages:
    with st.chat_message(message["role"]):
        st.markdown(message["content"])

def chat_flow():
    user_message = st.session_state.prompt_input

    if not user_message:
        return

    # 1. Add user message to state
    st.session_state.messages.append({"role": "user", "content": user_message})
    with st.chat_message("user"):
        st.markdown(user_message)

    # 2. Get response from agent
    with st.spinner('Thinking...'):
        response_text, tools_used = st.session_state.agent.process_message(user_message)

    # 3. Format response text
    if tools_used:
        tool_names = [t.replace('_', ' ').title() for t in tools_used]
        tool_names_display = ', '.join(set(tool_names))
        response_text += f"\n\n---\n**üõ†Ô∏è Tools Used:** {tool_names_display}"

    # 4. Display assistant response
    with st.chat_message("assistant"):
        st.markdown(response_text)

    # 5. Add assistant response to state
    st.session_state.messages.append({"role": "assistant", "content": response_text})

    # Clear the temporary input state variables
    del st.session_state.prompt_input
    if 'run_chat' in st.session_state:
        del st.session_state.run_chat


if 'run_chat' in st.session_state and st.session_state.run_chat:
    chat_flow()


# Chat Input box
if 'prompt_input' not in st.session_state:
     st.session_state.prompt_input = ""
prompt = st.chat_input("Share what's on your mind... I'm here to help.", key="chat_input")

if prompt:
    st.session_state.prompt_input = prompt
    chat_flow()
    st.rerun() # Rerun to display the output immediately


# --- Sidebar/Footer for Reset and Info ---

def reset_chat():
    st.session_state.agent.reset()
    st.session_state.messages = []
    st.rerun()

st.sidebar.markdown("## ‚ÑπÔ∏è Assistant Control")

if st.sidebar.button("üîÑ Start New Conversation", type="secondary", use_container_width=True):
    reset_chat()

st.sidebar.markdown("""
---
**Architecture:** Google Agent Development Kit (ADK) pattern
**Model:** Gemini 2.5 Flash (`gemini-2.5-flash`)

**Available Tools:**
* ü´Å Breathing exercises (short/medium/long)
* üßò Grounding techniques (5-4-3-2-1, physical, reality orientation)
* üÜò Emergency resources (India, USA, UK)
* üìã Evidence-based coping strategies
""")
```
